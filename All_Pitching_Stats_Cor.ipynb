```python
# All imports required for this project
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python
# Load the dataset
df = pd.read_csv('/Users/evansmith/Desktop/Pineville Porcupine Stats/Pitcher Only Stats Until Jul 3.csv')
```

```python
# Replaces any stats with a -, with a 0
df = df.replace("-", 0)
df.tail()
```

# FIP Correlations (It is basically an advanced ERA metric)

```python
# Calculating the correlation and p-value between FIP and strike percentage
corr, p_value = pearsonr(df["FIP"], df["S%"])

print(f"Correlation: {corr:.3f}")
print(f"P-value: {p_value:.10f}")
```

```python
# Plots the calculated correlation and shows the line of best fit
sns.regplot(data=df, x="FIP", y="S%")

plt.title("Fielding Independent Pitching vs Strike Percentage")
plt.xlabel("Fielding Independent Pitching (FIP)")
plt.ylabel("Strike Percentage (S%)")

plt.show()
```

```python
# Calculating the correlation and p-value between FIP and walks plus hits per inning
corr, p_value = pearsonr(df["FIP"], df["WHIP"])

print(f"Correlation: {corr:.3f}")
print(f"P-value: {p_value:.10f}")
```

```python
# Plots the calculated correlation and shows the line of best fit
sns.regplot(data=df, x="FIP", y="WHIP")

plt.title("Fielding Independent Pitching vs Walks per Inning")
plt.xlabel("Fielding Independent Pitching (FIP)")
plt.ylabel("Walks per Inning (BB/INN)")

plt.show()
```

```python
# Calculating the correlation and p-value between FIP and runs
corr, p_value = pearsonr(df["FIP"], df["R"])

print(f"Correlation: {corr:.3f}")
print(f"P-value: {p_value:.10f}")
```

```python
# Plots the calculated correlation and shows the line of best fit
sns.regplot(data=df, x="FIP", y="R")

plt.title("Fielding Independent Pitching vs Runs")
plt.xlabel("Fielding Independent Pitching (FIP)")
plt.ylabel("Runs (R)")

plt.show()
```

# Other Intersting Finds

```python
# Calculating the correlation and p-value between loses and leadoff walks
corr, p_value = pearsonr(df["L"], df["LOBB"])

print(f"Correlation: {corr:.3f}")
print(f"P-value: {p_value:.10f}")
```

```python
# Plots the calculated correlation and shows the line of best fit
sns.regplot(data=df, x="L", y="LOBB")

plt.title("Loses vs Leadoff walk (1st batter of inning)")
plt.xlabel("Loses")
plt.ylabel("Leadoff walk (1st batter of inning)")

plt.show()
```

### This shows that giving up walks early likely lead to loses for pitchers
